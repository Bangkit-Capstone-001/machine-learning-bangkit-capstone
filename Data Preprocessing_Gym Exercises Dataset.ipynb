```python
import pandas as pd
```

```python
df = pd.read_csv('Gym Exercises Dataset.csv')
```

```python
df.head(5)
```

```python
#Checking data type
df.info()
```

```python
#Dropping unused coloumns
df = df.drop(columns=['Exercise_Image','thumbnail_url', 'gambar thumbnail yt', 'Valid?', 'Unnamed: 13'])
```

```python
#Checking Missing Value
df.isnull().sum()
```

There is a missing value in every column

```python
#Checking the location of missing values in each column
missing_values = df[df.isnull().any(axis=1)]
missing_indices = df[df.isnull().any(axis=1)].index
print("Rows Contain Missing Value:", missing_indices.tolist())
```

It turns out the missing value is in the last row, so let's just drop that row.

```python
#dropping row contains missing value
df=df.dropna()
```

```python
#Re-checking missing value
df.isnull().sum()
```

```python
# Checking duplicates based on the Exercise_Name column.
duplicates = df.duplicated(subset=['Exercise_Name'])
```

```python
#dropping duplicates data
df_cleaned = df.drop_duplicates(subset=['Exercise_Name'], keep='first')
```

```python
df_cleaned
```

```python
#Write the DataFrame to an Excel file
df_cleaned.to_excel('Full Gym Equipment.xlsx', index=False)
```

```python
#Subsetting the data based on the Equipment column.
df1 = df_cleaned[df_cleaned['Equipment'].isin(['Body Only', 'Weight Bench'])]
df2 = df_cleaned[df_cleaned['Equipment'].isin(['Body Only','Dumbbell', 'Dumbbells', 'Kettlebells'])]
```

```python
#Write the df1 and df2 to Excel files
df1.to_excel('Body Only.xlsx', index=False)
df2.to_excel('Minimal Equipments.xlsx', index=False)
```
